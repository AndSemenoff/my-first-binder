## Пример 1. График курса рубля

Напишем простой пример считывание данных о курсе рубля с сервиса yahoo finance и построим простой линейный график. Используем библиотеку yfinance

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import yfinance as yf
FONT_SIZE = 16
```

```python
# В yahoo у рубля обозначение "RUB=X"
rub = yf.Ticker("RUB=X")
mstart = "2018-06-13" # yyyy-mm-dd
mend = "2022-01-07"
rub_historical = rub.history(start = mstart, end = mend, interval="1d")

fig = plt.figure(figsize=(16, 8))

ax1 = fig.subplots(1)
ax1.plot(rub_historical.Close)
ax1.set_title(f"Курс рубля от {start} до {end}", size=FONT_SIZE+6, color='black')
ax1.set_xlabel("Дата", size=FONT_SIZE, color='black')
ax1.set_ylabel("Цена закрытия", size=FONT_SIZE, color='black')
ax1.grid(True)
plt.tight_layout() # автоматическое выравнивание элементов на холсте plt
```
